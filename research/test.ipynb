```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap  = 100,
)
```

```python
texts = text_splitter.split_text(raw_text)
```

```python
len(texts)
```

```python
texts[0]
```

```python
from langchain_core.documents import Document
```

```python
docs = []
for i in range(len(texts)):
    doc = Document(page_content=texts[i])
    docs.append(doc)
```

```python
docs
```

```python
vstore.add_documents(docs)
```

```python
retriever = vstore.as_retriever(search_kwargs={"k": 3})
```

```python
retriever.get_relevant_documents("what is Market For Registrant’s Common Equity?")
```

```python
PRODUCT_BOT_TEMPLATE = """
    You are an expert at all things finance.
    Ensure your answers are relevant to the query context and refrain from straying off-topic.
    Your responses should be concise and informative.

    CONTEXT:
    {context}

    QUESTION: {question}

    YOUR ANSWER:
    
    """
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_template(PRODUCT_BOT_TEMPLATE)
```

```python
from langchain_huggingface import HuggingFacePipeline
from huggingface_hub import login
login(token=HF_TOKEN)
model_id = "meta-llama/Llama-3.2-1B-Instruct"
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto"
)

tokenizer = AutoTokenizer.from_pretrained(model_id)

pipeline = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
        use_cache=True,
        device_map="auto",
        max_new_tokens=500,
        do_sample=True,
        top_k=5,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.eos_token_id,
)
```

```python
llm = HuggingFacePipeline(pipeline=pipeline)
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
chain = (
        {"context": retriever, "question": RunnablePassthrough()}
        | prompt
        | llm
        | StrOutputParser()
    )
```

```python
chain.invoke("what is Market For Registrant’s Common Equity?")
```
